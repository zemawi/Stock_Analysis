##### **Correlation between Volume indicators(OBV, MFI, AD), sentiment, price and Daily return**

```python
corr_analysis.correlate_indicator_returns('volume')
```

**Observation**
- The above analysis focus on the relaionship between volume indicators and sentiment of news with respect to daily returns.
    * **Findings**: The Close price has a very strong positive correlation with OBV (On-Balance Volume) and AD (Accumulation/Distribution), both with correlation coefficients of 0.94 and 0.96, respectively. This suggests that as the Close price increases, these indicators tend to increase as well, indicating a strong relationship between the price and volume-based indicators.

    * **Findings**: AD also has a strong positive correlation with OBV (0.90), which makes sense since both indicators are volume-based and measure similar market dynamics, though with different methodologies.

    * **Findings**: The MFI (Money Flow Index) has a moderate positive correlation with sentiment (0.33), indicating that as sentiment improves, the MFI tends to increase. This might imply that positive news sentiment could lead to increased buying pressure, reflected in the MFI.

    * **Findings**: The Daily returns show only weak correlations with most of the other indicators, with the highest correlation being with OBV (0.25). This suggests that daily returns are not strongly influenced by the indicators analyzed here, though OBV seems to have some influence.

    * **Findings**: The plot of Sentiment versus MFI (Money Flow Index) reveals that these two indicators tend to move in the same direction. This suggests that as sentiment improves, indicating a more positive outlook, the MFI also increases, reflecting greater buying pressure in the market. However, the alignment between these indicators appears to exhibit a lag, which could be attributed to the market behavior of "buying the rumor and selling the fact/news." This means that investors might act on positive sentiment before the news is fully realeased, leading to a delay in the MFI's response to changes in sentiment. <u>From this, the MFI indicator might predict how the market is responding prior to the release of news.</u>
